```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
hlth_data = pd.read_stata("finaldata_hlthdemo.dta")
covid_data = pd.read_excel("Covid_data_17Feb2022.xlsx")
```

```python
hlth_data.head()
```

```python
covid_data.columns=["index", "Country", "Total_cases", "Total_death", "Total_recovered", "Active_cases", "Serious_Critical", "Total_cases_1m_pop", "Deaths_1m_pop", "Total_tests", "Tests_1m_pop", "Population"]
covid_data.head()
```

```python
# hlth_test = hlth_data.loc[(hlth_data["country"] == "United States") & (hlth_data["year"] == 2002)]
hlth_test = hlth_data.loc[(hlth_data["year"] == 2002)]
hlth_test["Country"] = hlth_data["country"]
result = pd.concat([covid_data, hlth_test], axis=1)
result["country"]
```

```python
merged = pd.merge(covid_data, hlth_test, on="Country")
merged.head()
```

```python
merged.info()
```

```python
reg = smf.ols("Total_cases_1m_pop  ~ democracy + lifeexp + sanitation", data=merged)
results = reg.fit()
print(results.summary())
```

---
First I wanted to test whether total cases were related to whether a country is democratic and/or has a good healthcare.

To start I run a regression with total cases per 1 million, as our dependent, on a (binary) democracy variable, life expectancy and (indexed) sanitation variable. Total cases per 1 million was chosen to avoid, bias towards countries with higher populations. From the results of the regression you can see that all variables are significant at the 95% confidence level. The coefficients do give us some strange results however. Whether or not a country is a democracy seems to increase the number of cases per 1 million by almost 30,000. Life expectancy and sanitation also give way to higher total cases per 1 million. This is a somewhat counter-intuitive reuslt, at first. However, most of the countries in the merged dataset are likely democratic, and many of the wealthier (and healthier) nations probably have higher populations in general and have better reporting systems. I should also note that the 3 independent variables are also from the year 2002 for each available country.

```python
reg2 = smf.ols("Serious_Critical  ~ democracy + lifeexp + sanitation", data=merged)
results2 = reg2.fit()
print(results2.summary())
```

---
My second regression checks to see if the number of Serious or Critical cases are any different in democratic and/or healthy nations. The results above tell us the a democratic country tends to see higher levels of serious covid cases, however this result is insignificant. Life expectancy also trends positively with serious cases, and is significant at the 95% confidence level. More intuitively we see that sanitation tends to lower the level of serious cases in a country. The former two variables tend to have counter-intuitive results again, but this is likely due to the bias in the number of democratic countries in the sample.
